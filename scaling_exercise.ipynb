```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pydataset
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler, QuantileTransformer

import prepare_telco
import wrangle
import os
from env import user, password, host
```

```python
df=wrangle.wrangle_zillow()
```

```python
#min max scaler
#split data
```

```python
train, validate, test = prepare_telco.split(df, stratify_by=None)
```

```python
train.shape, validate.shape, test.shape
```

```python
scaler = MinMaxScaler()
# Note that we only call .fit with the TRAINING data,
# but we use .transform to apply the scaling to all the data splits.
scaler.fit(train)

train_scaled = scaler.transform(train)
validate_scaled = scaler.transform(validate)
test_scaled = scaler.transform(test)

plt.figure(figsize=(15, 8))
plt.subplot(121)
plt.hist(train, bins=15, ec='black')
plt.title('Original')
plt.legend(train)

plt.subplot(122)
plt.hist(train_scaled, bins=15, ec='black')
plt.title('Scaled')
plt.legend(train);
```

```python
#min max inverse
```

```python
train_scaled = scaler.inverse_transform(train)
validate_scaled = scaler.inverse_transform(validate)
test_scaled = scaler.inverse_transform(test)

plt.figure(figsize=(15, 6))
plt.subplot(121)
plt.hist(train, bins=15, ec='black')
plt.title('Original')
plt.legend(train)

plt.subplot(122)
plt.hist(train_scaled, bins=15, ec='black')
plt.title('Scaled')
plt.legend(train)
plt.show()
```

```python
def scaled_inverseminmax(train, validate, test):
    scaler = MinMaxScaler()
    scaler.fit(train)
    train_scaled = scaler.inverse_transform(train)
    validate_scaled = scaler.inverse_transform(validate)
    test_scaled = scaler.inverse_transform(test)
    return train_scaled, validate_scaled, test_scaled
```

```python
train_scaled, validate_scaled, test_scaled = scaled_inverseminmax(train, validate, test)
```
